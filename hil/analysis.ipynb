### Call JAVA API

```python
# graphql api call localhost:8080 /ideas
import requests
import json

# Define the GraphQL query
query = """
query { 
    ideas {
        name
        description
        author {
            firstName
            lastName
        }
    }
}
"""
# Define the URL for the GraphQL API
url = "http://localhost:8080"
# Define the headers for the request
headers = {
    "Content-Type": "application/json",
    "Accept": "application/json",
}
```

```python

# Send the request to the GraphQL API
response = requests.post(url, json={"query": query}, headers=headers)
# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    # Print the data in a readable format
    print(json.dumps(data, indent=4))
else:
    # Print the error message if the request failed
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
#define rest api query from graphql
query_rest = json.dumps(query)

# Send the request to the REST API
response_rest = requests.get(url+"/api/ideas/all", headers=headers)
# Check if the request was successful
if response_rest.status_code == 200:
    # Parse the JSON response
    data_rest = response_rest.json()
    # Print the data in a readable format
    print(json.dumps(data_rest, indent=4))
else:
    # Print the error message if the request failed
    print(f"Error: {response_rest.status_code}")
    print(response_rest.text)
```
